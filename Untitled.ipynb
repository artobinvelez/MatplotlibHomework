```python

%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "Data/city_data.csv"
ride_data_to_load = "Data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
city_ride_df = pd.merge(city_df, ride_df, on=["city", "city"])


# Display the data table for preview
city_ride_df.head()
```

```python
# Obtain the x and y coordinates for each of the three city types
urban = city_ride_df[city_ride_df["type"] == "Urban"].groupby([city_ride_df["city"]])
rural = city_ride_df[city_ride_df["type"] == "Rural"].groupby([city_ride_df["city"]])
suburban = city_ride_df[city_ride_df["type"] == "Suburban"].groupby([city_ride_df["city"]])

x_axis_u = urban["ride_id"].count()
x_axis_r = rural["ride_id"].count()
x_axis_s = suburban["ride_id"].count()


y_axis_u = urban["fare"].mean()
y_axis_r = rural["fare"].mean()
y_axis_s = suburban["fare"].mean()

z_axis_u = urban["driver_count"].mean()
z_axis_r = rural["driver_count"].mean()
z_axis_s = suburban["driver_count"].mean()

# Build the scatter plots for each city types
plt.scatter(x_axis_u, y_axis_u, label = "Urban", s=z_axis_u*5, color=["coral"], edgecolor="black", alpha = 0.75, marker="o")
plt.scatter(x_axis_r, y_axis_r, label = "Rural", s=z_axis_r*5, color=["gold"], edgecolor="black", alpha = 0.75, marker="o")
plt.scatter(x_axis_s, y_axis_s, label = "Suburban", s=z_axis_s*5, color=["lightskyblue"], edgecolor="black", alpha = 0.75, marker="o")
plt.grid()
# Add title and axis labels
plt.title("Pyber Ride Sharing Data 2016")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
# Create legend
plt.legend(fontsize = 10, title= "City Types", loc="best") 
# Make a note to denote the s axis
plt.text(42,35,"Note: Circle size correlates with driver count per city.")
# Save Figure
plt.savefig('bubble_pyber.png')
```

```python

# Show plot
plt.show()
```

```python
# Calculate Type Percents
fare_pers = city_ride_df.groupby(['type'])['fare'].sum()
fare_pers.head()
```

```python
# Build Pie Chart
labels = ["Rural","Suburban","Urban"]
explode = (0, 0, 0.1)
colors = ["gold", "lightskyblue", "lightcoral"]
plt.pie(fare_pers, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Fares by City Type")
# Save Figure
plt.savefig('per_fares_pie.png')
```
